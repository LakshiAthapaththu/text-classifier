```python
##read the train data corpus
import pandas as pd
import xgboost as xgb
data = pd.read_csv('/home/dinushi/Desktop/ML/TheCompleteCorpus/TrainingCorpus.csv')
data
```

```python
#convert dataframe to matrix
conv_arr= data.values

import numpy as np
#split matrix into 2 columns each into 1d array

arr2 = np.delete(conv_arr,[0],axis=1) #get lables column only
label_array = arr2.ravel() #create an array of labels
labels=np.array(label_array, dtype=np.float) #conver int to float

labels #the labels of the training data
```

```python
#create the bag of word matrix for train text corpus
corpus = list(data.Text) #the training text array

#the whole set of words in the corpus is collected
word_array = []
for sentence in corpus:
    #print (sentence.split())
    word_array.extend(sentence.split())
word_array = (set(word_array))
word_array= list(word_array)
#print(word_array)
```

```python
#create representation of each train text sentence  
sentence_list = []
for sentence in corpus:
    bag = [0]*len(word_array)
    for word in (sentence.split()):
        if(word in word_array):
           bag[(word_array.index(word))] = 1
    sentence_list.append(bag)
    
#the vector representation of each train text sentence     
sentence_list
```

```python
from numpy import array, matrix
from scipy.sparse import csr_matrix

#convert the vector array to a sparse matrix
X_train_counts = csr_matrix(sentence_list)
X_train_counts.shape
```

```python
#find Tf_idf value vector for the each sentence in training set
from sklearn.feature_extraction.text import TfidfTransformer

tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_train_tfidf.shape #print the size of the matrix with tf-idf value of each word in the sentence
```

```python
from sklearn.naive_bayes import MultinomialNB

#train the SVM model
#clf = SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, n_iter=5, random_state=42).fit(X_train_tfidf, labels)
#clf = SGDClassifier().fit(X_train_tfidf, labels)
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.185, learning_rate = 0.1591, max_depth = 5, alpha = 0.01, n_estimators = 22)
clf =xg_reg.fit(X_train_tfidf, labels)
#clf = MultinomialNB().fit(X_train_tfidf, labels)
```

```python
#load test data
data_test = pd.read_csv('/home/dinushi/Desktop/ML/testset.csv')
data_test
#convert dataframe to matrix
conv_arr2= data_test.values

#split matrix into 2 columns each into 1d array
arr3 = np.delete(conv_arr2,[0],axis=1) 
label_array_t = arr3.ravel()
labels_test=np.array(label_array_t, dtype=np.float)
labels_test #labels of test data
```

```python
corpus_test = list(data_test.Text)

#represent each test sentence as a vector
sentence_list2 = []
for sentence in corpus_test:
    bag = [0]*len(word_array)
    for word in (sentence.split()):
        if(word in word_array):
           bag[(word_array.index(word))] = 1
    sentence_list2.append(bag)
        
sentence_list2 #the vector representation of each test sentence

X_train_counts_test = csr_matrix(sentence_list2)

tfidf_transformer = TfidfTransformer()
X_train_tfidf_test = tfidf_transformer.fit_transform(X_train_counts_test)
X_train_tfidf_test.shape #print the size of the matrix with tf-idf value of each test sentence
```

```python
#predict the values for test data
predicted_svm = clf.predict(X_train_tfidf_test)
predicted_svm = predicted_svm.astype(int)

#compare with test labels
np.mean(predicted_svm == labels_test)
```
